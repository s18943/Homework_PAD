# c.
###	Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply

```python
# zamieniam wszystkie kolumny do odpowiedmnich typów 
df_ord.tshirt_category = df_ord.tshirt_category.astype('string')
df_ord.order_id = df_ord.tshirt_category.astype('string')
df_ord.customer_id = df_ord.customer_id.astype('string')
```

```python
t_shirt_dict = {'White T-Shirt M':'White T-Shirt M', 
                'White T-Shirt F':'White T-Shirt F', 
                'Wh Tshirt F':'White T-Shirt F',
       'Black T-Shirt F':'Black T-Shirt F',
        'Hoodie':'No Hoodie No', 
        'Black T-Shirt M':'Black T-Shirt M', 
        'Tennis Shirt':'No TennisShirt No',
       'Wh Tshirt M':'White T-Shirt M', 
       'Bl Tshirt F':'Black T-Shirt F', 
       'Bl Tshirt M':'Black T-Shirt M'}

df_ord.tshirt_category = df_ord.tshirt_category.apply(lambda x: t_shirt_dict[x])
df_ord.tshirt_category.unique()
```

# d.	
#### Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
df_ord[['tshirt_colour', 'tshirt_type', 'tshirt_sex']] = df_ord.tshirt_category.str.split(' ', 2, expand=True)
```
